## Tensor的维度和形状

请问上面定义的`vector`的维度是多少？

从不同角度出发可能有不同的答案。

`vector`作为一个向量，它的长度为5，因此我们可以说它是一个5维**向量**。

但是`vector`的存储结构是1维数组，因此我们可以说它是一个1维**张量**。

为了避免混淆，在张量运算中，我们一般讨论张量的形状。

```python
vector.shape
```

```python
matrix.shape
```

```python
transposable_vector.shape
```

```python
image_rgb.shape
```

可以看到：

- `vector`的形状是`(5,)`，这说明它只有1个维度，且这个维度上的长度是5，根据形状可以推断它是一个长度为5的向量
- `matrix`的形状是`(3, 3)`，这说明它有2个维度，两个维度上的长度均为3，根据形状可以推断它是一个3行3列的方形矩阵
- `transposable_vector`的形状是`(1, 5)`，根据形状可以推断它是一个1行5列的矩阵，即长度为5的行向量
- `image_rbg`的形状是`(2, 2, 3)`，根据形状和经验可以推断它是一张拥有2x2个像素点的图片，每个像素点拥有rgb三个值

## 向量的内积

向量和向量可以计算内积：

```python
a = np.array([1, 2, 3])
b = np.array([3, 2, 1])
a @ b
```

严谨一点也可以写成一个行向量和一个列向量的内积：

```python
a_row = np.array([[1, 2, 3]])
b_col = np.array([
    [3],
    [2],
    [1],
])
a_row @ b_col
```

注意，这两种方式虽然都能得到答案10。

但是，前者得到的是一个标量数值，后者得到的是一个形状为`(1, 1)`的tensor

因为后者其实是一个形状为`(1, 3)`的矩阵与一个形状为`(3, 1)`的矩阵相乘得到一个形状为`(1, 1)`的矩阵

另外，前者是单纯的向量内积，满足交换律：

```python
b @ a
```

后者本质上是矩阵乘法，不满足交换律，交换后变成了`(3, 1)`矩阵乘`(1, 3)`矩阵得到`(3, 3)`矩阵：

```python
b_col @ a_row
```

## 矩阵乘法

```python
mat_A = np.array([
    [0, 1],
    [1, 0],
    [0, 1],
    [1, 1]
])
mat_B = np.array([
    [1, 0, 1],
    [0, 1, 1],
])
mat_A @ mat_B
```

在你的代码需要进行矩阵乘法的时候，最好先检查参与相乘的两个矩阵的形状。这样可以保证矩阵乘法顺利进行。

```python
mat_A.shape
```

```python
mat_B.shape
```

`mat_A`第2维的长度（列数）必须与`mat_B`第1维的长度（行数）相同，两个矩阵才能相乘

# Self-Attention计算过程

2024-11-26-14-29-10.png

```python
import numpy as np
from scipy.special import softmax
```

输入为$X=(\boldsymbol{a}^1,\boldsymbol{a}^2,\boldsymbol{a}^3,\boldsymbol{a}^4)$，其中$\boldsymbol{a}^i$为列向量

输入序列的长度为$|X|=4$

假设输入序列中每个输入向量的长度为$H=|\boldsymbol{a}^i|=3$

可以用代码定义$X$如下：

```python
X = np.array([
    [1.0, 0.0, 1.0],  # a1
    [0.0, 2.0, 0.0],  # a2
    [1.0, 1.0, 0.0],  # a3
    [0.0, 1.0, 2.0],  # a4
])
```

注意！李宏毅的ppt中$\mathbf{a}^i$是列向量，但是代码定义中$\mathbf{a}^i$是行向量。

代码中一般都将输入向量定义为行向量，因为这样可读性更好。

$W^q$应该是一个大小为$3 \times 3$的方形矩阵，因为输入$X$中每个向量的长度$|\mathbf{a}^i|=3$，输出的$\mathbf{q}^i$长度应该与$\mathbf{a}^i$相同，也为3

```python
W_q = np.array([
    [1.0, 0.0, 0.1],
    [0.1, 0.5, 0.0],
    [0.2, 0.1, 1.0],
])
```

$W^k$和$W^v$同理：

```python
W_k = np.array([
    [0.0, 0.5, 1.0],
    [0.1, 0.0, 0.2],
    [1.0, 0.5, 0.1],
])
W_v = np.array([
    [1.0, 0.5, 0.1],
    [0.1, 0.5, 0.5],
    [1.0, 0.0, 0.0],
])
```

我们先用代码计算$\mathbf{a}^1$与$\mathbf{a}^2$的attention score，这一过程的示意图如下：

2024-11-26-09-27-42.png

在李宏毅ppt的定义中，是拿$W^Q$的每一行（总共3行）乘上列向量$\boldsymbol{a}^1$，得到列向量$\boldsymbol{q}^1$，如下图所示：

2024-11-26-10-03-35.png![](screenshots/2024-11-26-10-03-35.png)

在我们的代码中，`a1`和`q1`都是行向量，因此应该拿行向量`a1`乘上$W^Q$的每一列（总共3行），得到行向量`q1`：

2024-11-26-12-52-49.png![](screenshots/2024-11-26-12-52-49.png)

```python
a1 = X[0]
q1 = a1 @ W_q
```

```python
q1
```

同理可以计算$\boldsymbol{k}^2$:

```python
a2 = X[1]
k2 = a2 @ W_k
```

```python
k2
```

将$\boldsymbol{q}^1$和$\boldsymbol{k}^2$作内积，得到$\alpha_{1,2}$:

```python
alpha_1_2 = q1 @ k2
```

可得$\boldsymbol{a}^1$与$\boldsymbol{a}^2$的attention score $\alpha_{1,2}=0.68$

```python
alpha_1_2
```

同理可以计算任意$\boldsymbol{a}_i$和$\boldsymbol{a}_j$之间的attention score $\alpha_{i,j}$：

$$
\alpha_{i,j}=\boldsymbol{q}^i\cdot\boldsymbol{k}^j
$$

为了加速运算，我们可以把所有的**行**向量$\mathbf{q}^i$拼起来形成一个矩阵$Q$：

（李宏毅的投影片是把$\mathbf{q}^i$看作列向量拼起来的，和原论文不一致，这里的做法和原论文才是一致的）

2024-11-26-12-45-39.png![](screenshots/2024-11-26-12-45-39.png)

所有的**行**向量$\mathbf{k}^j$也拼起来形成一个矩阵$K$：

2024-11-26-12-45-15.png![](screenshots/2024-11-26-12-45-15.png)

把$Q$乘上$K^T$，就能得到一个方形矩阵$A$，这个矩阵的行数列数均为输入序列的长度（token个数），矩阵中$(i,j)$个位置的值表示输入向量$\boldsymbol{a}_i$和$\boldsymbol{a}_j$之间的attention score:

（李宏毅的这个attention score矩阵正好是转置的）

2024-11-26-12-55-46.png![](screenshots/2024-11-26-12-55-46.png)

代码实现如下：

计算`q1`到`q4`然后拼成矩阵`Q`:

```python
Q = []
for i in range(4):
    ai = X[i]
    qi = ai @ W_q
    Q.append(qi)
Q = np.array(Q)
```

```python
Q
```

计算`k1`到`k4`然后拼成`K`:

```python
K = []
for i in range(4):
    ai = X[i]
    ki = ai @ W_k
    K.append(ki)
K = np.array(K)
```

```python
K
```

计算attention score矩阵：

```python
Alpha = Q @ K.T
```

```python
Alpha
```

事实上`q1`到`q4`以及`k1`到`k4`也没有必要在循环里一个一个算完再拼起来

完全可以用矩阵乘法并行地算：

2024-11-26-12-57-38.png![](screenshots/2024-11-26-12-57-38.png)

```python
Q = X @ W_q
K = X @ W_k
Alpha = Q @ K.T
```

```python
Alpha
```

```python
Alpha_prime = softmax(Alpha, axis=-1)
```

```python
Alpha_prime
```

跟计算$Q,V$同理，我们可以算出$V=(v^1,v^2,v^3,v^4)$:

```python
V = X @ W_v
```

$\boldsymbol{b}^1$是$v_1$到$v_4$的加权和，权重为$\alpha'_{1,1}$到$\alpha'_{1,4}$：

2024-11-26-12-29-48.png![](screenshots/2024-11-26-12-29-48.png)

可以简单地用代码实现为：

```python
b1 = (
    Alpha_prime[0, 0] * V[0] +
    Alpha_prime[0, 1] * V[1] +
    Alpha_prime[0, 2] * V[2] +
    Alpha_prime[0, 3] * V[3]
)
b1
```

也可以并行地计算$b_1$到$b_4$。

用经过softmax以后的attention score矩阵$A'$乘上矩阵$V$，计算过程如下：

2024-11-26-14-57-07.png![](screenshots/2024-11-26-14-57-07.png)

每个向量$\boldsymbol{v}^i$长度为3，也就是说有3个位置。向量的加权和就是对对应位置上的值作加权和。

$A'$与$V$相乘时就是在计算对应位置上的值的加权和。以$A'$的第1行与$V$中的每一列相乘为例：

- 与第1列相乘：对4个向量的第1个位置上的值$\boldsymbol{v}^1_1,\boldsymbol{v}^2_1,\boldsymbol{v}^3_1,\boldsymbol{v}^4_1,$求加权和
- 与第2列相乘：对4个向量的第2个位置上的值$\boldsymbol{v}^1_2,\boldsymbol{v}^2_2,\boldsymbol{v}^3_2,\boldsymbol{v}^4_2,$求加权和
- 与第3列相乘：对4个向量的第3个位置上的值$\boldsymbol{v}^1_3,\boldsymbol{v}^2_3,\boldsymbol{v}^3_3,\boldsymbol{v}^4_3,$求加权和

相乘产生的结果又形成了一个向量$b_1$，$b_1$就是以$A'$中的第1行为权重，$\boldsymbol{v}^1,\boldsymbol{v}^2,\boldsymbol{v}^3,\boldsymbol{v}^4$的加权和

同理可得$b_2,b_3,b_4$.

通过$A'$和$V$计算输出的代码如下：

```python
Y = Alpha_prime @ V
```

```python
Y[1]  # b2
```

```python
Y[2]  # b3
```

```python
Y[3]  # b4
```

## 总结

给定输入$X=(a_1,a_2,a_3,a_4)$，得到输出$Y=(b_1,b_2,b_3,b_4)$的self-attention机制可以定义为以下函数：

```python
def self_attention(X):
    W_q = np.array([
        [1.0, 0.0, 0.1],
        [0.1, 0.5, 0.0],
        [0.2, 0.1, 1.0],
    ])
    W_k = np.array([
        [0.0, 0.5, 1.0],
        [0.1, 0.0, 0.2],
        [1.0, 0.5, 0.1],
    ])
    W_v = np.array([
        [1.0, 0.5, 0.1],
        [0.1, 0.5, 0.5],
        [1.0, 0.0, 0.0],
    ])
    Q = X @ W_q
    K = X @ W_k
    V = X @ W_v
    attention_scores = softmax(Q @ K.T, axis=-1)
    Y = attention_scores @ V
    return Y
```

```python
self_attention(X)
```
